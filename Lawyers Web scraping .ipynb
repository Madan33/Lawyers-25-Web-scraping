## Put Everything Hear inside  for-loop

```python
Name = []
Short_Bio = []
Specialization = []
university = []
Address = []
Phone = []
Email_link = []

for result in results:
    # Name
    try:
        Name.append(result.find("strong", {"class" : "lawyer-name"}).get_text(strip=True))
    except:
        Name.append(" ")
    
    # Short_Bio
    try : 
        Short_Bio.append(result.find("div", {"class":"lawyer-expl"}).get_text(strip=True))
    except:
        Short_Bio.append(" ")

    #Specialization
    try :
        Specialization.append(result.find("span", {"class":"-practices"}).get_text(strip=True))
    except:
        Specialization.append(" ")
    
    # university
    try : 
        university.append(result.find("span", {"class":"-law-schools"}).get_text())
    except:
        university.append(" ")
    
    #Address
    try :
         Address.append(result.find("span",{"class":"-address"}).get_text(strip=True).replace("\t", " "))
    except:
        Address.append(" ")
     
    #Phone 
    try :
        Phone.append(result.find("strong",{"class":"-phone"}).get_text(strip=True))
    except:
        Phone.append(" ")


   # Email_link
    try:
        Email_link.append(result.find("a",{"class":"-email"}).get("href"))
    except:
        Email_link.append("")
```

## create Pandas Data frame 

```python
Df_Lawyers =pd.DataFrame({"layer_name": Name,"Short_Bio":Short_Bio,"Specialization":Specialization,"university":university,"Address":Address,"Phone":Phone,"email":Email_link})
```
